```python
def createAlly():
    pd.set_option('display.max_rows', None)  # Show all rows
    pd.set_option('display.max_columns', None)  # Show all columns
    
    input_file = "rawAlly.csv"
    output_file = "ally.csv"
    df=pd.read_csv(input_file)
    df=df.rename(columns={" Type":"TransactionType"," Amount":"Amount" ," Time":"Time"," Description":"Desc"})
    df["TransactionType"]=df["TransactionType"].replace({"Withdrawal":"Credit","Deposit":"Debit"})
    df.loc[:,"Date"]=df.loc[:,"Date"].astype(str)
    newRow=["7/30/2023","0:00:00",912.28,"Debit","Initial Balance"]
    df.loc[len(df)]=newRow
    df["Date"]=pd.to_datetime(df.loc[:,"Date"])
    df["Category"]="Reset"
    df["Source"]="Ally"
    ally=df[["Date","Desc","Amount","TransactionType","Category","Source"]]
    ally=ally.sort_values("Date",ascending=True)
    ally.to_csv(output_file,index=False)
    return ally
```

```python
def createEsl():
    # Read the CSV and modify the header
    input_file = "rawESL.csv"
    output_file = "esl.csv"
    category_file="categoryLookup.csv"
    
    with open(input_file, "r", newline="", encoding="utf-8") as infile, \
         open(output_file, "w", newline="", encoding="utf-8") as outfile:
    
        reader = csv.reader(infile)
        writer = csv.writer(outfile)
        
        # Skip the first 3 rows
        for _ in range(3):
            next(reader, None)
        # Copy the rest of the data
        for row in reader:
            writer.writerow(row)
    
    df=pd.read_csv(output_file)

    
    df = df.rename(columns={"Description":"Desc"})
    df = df.rename(columns={"combinedAmt":"Amount"})
    df["Amount Credit"]=df["Amount Credit"].fillna(0)
    df["Amount Debit"]=df["Amount Debit"].fillna(0)
    df["Amount"]=df["Amount Debit"] +df["Amount Credit"]
    df["TransactionType"] = df["Amount"].apply(lambda x: "Credit" if x > 0 else "Debit")
    df.fillna("", inplace=True)
    df["newDesc"]=df["Desc"]+df["Memo"]
    df=df[["Date","Amount","TransactionType","newDesc"]]
    new_row=["1/1/2024",1665.99,"Debit","Initial balance"]
    df.loc[len(df)]=new_row

    mask = df['newDesc'].str.contains('Rent', case=False)
    water_rows = df[mask].copy()
    water_rows['newDesc'] = 'Water Bill'
    water_rows['Amount'] = df[mask]['Amount'] + 845  # Water bill is remaining amount
    
    # Adjust rent transactions to be exactly $800
    df.loc[mask, 'Amount'] = -845
    df["Date"]=pd.to_datetime(df.loc[:,"Date"])
    df=df.sort_values(by="Date",ascending=True)
    df = df.rename(columns={"newDesc":"Desc"})
    df["Desc"]=df["Desc"].astype(str)
    df["Category"]="Reset"
    df["Source"]="ESL"
    esl=df
    esl.to_csv("esl2.csv")
    return df
```

```python
def createDiscover():
    input_file = "rawDiscover.csv"
    output_file = "discover.csv"
    
    df=pd.read_csv(input_file)
    df["TransactionType"]=df.loc[:,"Amount"].apply(lambda x:"Credit" if x>0 else "Debit")
    df["Description"]=df["Description"].astype(str)
    
    df=df.rename(columns={"Trans. Date":"Date"})
    df=df.rename(columns={"Description":"Desc"})
    df=df.rename(columns={"Category":"OringalCat"})
    df["Category"]=df["OringalCat"]
    df["Date"]=pd.to_datetime(df.loc[:,"Date"])
    df["Source"]="Discover"
    def setCategory(row):
        if "APPLE" in row["Desc"].upper() or "MEATBUS" in row["Desc"].upper() or "YNAB" in row["Desc"].upper():
            return "Subscription"
        elif "COMCAST" in row["Desc"].upper() or "PUGET SOUND ENERGY" in row["Desc"].upper() or "SANITARY SERVICE" in row["Desc"].upper():
            return "Utilities"
        elif "ABELSPEAKS" in row["Desc"].upper() or "WEGMANSEASTAVEN" in row["Desc"].upper():
            return "Charity"
        else:
            return row["OringalCat"]
    
    
    def setDesc(row):
        if row["Amount"] == 10.89 and "APPLE" in row["Desc"].upper():
            return "MAX"
        elif row["Amount"] == 2.99 and "APPLE" in row["Desc"].upper():
            return "iCloud Storage"
        else:
            return row["Desc"]
    
    df["Category"]=df.apply(setCategory,axis=1)
    df["Desc"]=df.apply(setDesc,axis=1)
    discover=df[["Date","Desc","Amount","TransactionType","Category","Source"]]
    discover.to_csv(output_file,index=False)
    return discover
```

```python
esl=createEsl()
discover=createDiscover()
ally=createAlly()

def setCategory(row):
    if "APPLE" in row["Desc"].upper() or "MEATBUS" in row["Desc"].upper() or "YNAB" in row["Desc"].upper():
        return "Subscription"
    elif "COMCAST" in row["Desc"].upper() or "PUGET SOUND ENERGY" in row["Desc"].upper() or "SANITARY SERVICE" in row["Desc"].upper():
        return "Utilities"
    elif "ABELSPEAKS" in row["Desc"].upper() or "WEGMANSEASTAVEN" in row["Desc"].upper():
        return "Charity"
    else:
        return row["Category"]

def setDesc(row):
    if row["Amount"] == 10.89 and "APPLE" in row["Desc"].upper():
        return "MAX"
    elif row["Amount"] == 2.99 and "APPLE" in row["Desc"].upper():
        return "iCloud Storage"
    else:
        return row["Desc"]


def categorize_descriptions(df,lookup_df,catStat_lookup_dict):
    """
    Maps description strings to standardized categories using a mapping dictionary
    and tracks changes in a 'changed' column.
    
    Args:
        df (pd.DataFrame): DataFrame containing a 'Desc' column
        
    Returns:
        pd.DataFrame: DataFrame with updated 'Desc' and 'changed' columns
    """

    
    # Create a copy of the DataFrame to avoid modifying the original
    df = df.copy()
    

    lookup_dict = dict(zip(lookup_df['Desc'].str.upper(), lookup_df['Category']))
    # Define mapping of search terms to categories
    catStat_lookup_dict=dict(zip(catStat_lookup_dict["Category"].str.upper(),catStat_lookup_dict["CategoryStatus"]))
    
    def map_category(row):
        if row['Desc']=="nan":
            if 'TransactionType' in row:
                if row['TransactionType'].upper() == 'DEBIT':
                    return 'Unknown Expense'
                elif row['TransactionType'].upper() == 'CREDIT':
                    return 'Unknown Income'
            return 'Unknown Transaction'
        
        desc_upper = row['Desc'].upper()
        original_desc = row['Desc']
        
        for search_term, category in desc_mapping.items():
            if search_term.upper() in desc_upper:
                return category
        return original_desc


    def match_category(row):
        desc_upper = row['Desc'].upper() if pd.notna(row['Desc']) else ''
        
        # Check for exact match first
        if desc_upper.upper() in lookup_dict:
            return lookup_dict[desc_upper.upper()]
        
        # Check for partial matches
        for lookup_desc, category in lookup_dict.items():
            if lookup_desc in desc_upper:
                return category
        
        # No match found
        return row["Category"]

    def match_CategoryStatus(row):
        category_upper = row['Category']
        
        # Check for exact match first
        if category_upper.upper() in catStat_lookup_dict:
            # print(catStat_lookup_dict[category_upper])
            return catStat_lookup_dict[category_upper.upper()]


    # Apply the mapping function to each row
    result = df.apply(map_category, axis=1)
    df['Desc'] = result    
    
    # Apply the matching function
    # result = df.apply(match_category, axis=1)
    df['Category'] = df.apply(match_category,axis=1)
    # result=df.apply(match_CategoryStatus,axis=1)
    # df["categoryStatus"]=result
    df["CategoryStatus"] = df.apply(match_CategoryStatus, axis=1).fillna(df.get("categoryStatus", "Unknown"))

    return df


    
catDF=pd.read_csv("Files/categoryLookup.csv")
catStatDF=pd.read_csv("Files/categoryStatusLookup.csv")
all=pd.concat([ally,esl,discover],ignore_index=True)



all=all.sort_values(by="Date",ascending=True)
all['id'] = range(1, len(all) + 1) 

all["Category"]=all.apply(setCategory,axis=1)
all= categorize_descriptions(all,catDF,catStatDF)
all['Week'] = all['Date'].dt.to_period('W')
weekly_balance = all.groupby('Week')['Amount'].sum().reset_index()
weekly_balance.to_csv("weekly_balance.csv",index=False)

mask = all[all['Desc'].str.contains('Rent')]
mask=mask[mask["Amount"]>845.00]
water_rows = all[mask].copy()
water_rows['Desc'] = 'Water Bill'
water_rows['Amount'] = all[mask]['Amount'] + 845.0  # Water bill is remaining amount
mask = all['Desc'].str.contains('Rent')
# Adjust rent transactions to be exactly $800
all.loc[mask, 'Amount'] = -845.0

all.to_csv("all.csv",index=False)
all['Balance'] = all.groupby('Source')['Amount'].cumsum()
all
```
